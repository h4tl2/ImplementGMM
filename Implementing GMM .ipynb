### Data  
2-dimensional from 5 gaussian mixtures

```python
#hardcode fix the K 
def gen_dataset():
    #hardcode the mean
    mean = [[3,5],[9,2],[15,7],[20,3],[24,5]]
    cov = [[1, 0], [0, 1]]
    x = np.zeros((0, 2), dtype=np.float32)
    for i in range(5):
        x_temp = np.random.multivariate_normal(mean[i], cov, 100)
        x = np.vstack((x,x_temp))
    return x
```

```python
def gmm1(x,k=5,iterations=100):
    #initial mu and sigma from some data point by random
    n,d = x.shape # check number of the data
    mu = x[np.random.choice(n, k, False), :] 
    initial_mu = mu[:]
    sigma= [np.eye(d)] * k
    initial_sigma = sigma[:]
    w = [1./k] * k #weight as define in the equation above
    R = np.zeros((n, k)) #responsbility matrix using in e step calculate from the current param
    log_likelihoods = []
    #define pdf 
    P = lambda mu, s: np.linalg.det(s) ** -.5 ** (2 * np.pi) ** (-x.shape[1]/2.) \
                * np.exp(-.5 * np.einsum('ij, ij -> i',\
                        x - mu, np.dot(np.linalg.inv(s) , (x - mu).T).T ) ) 


    #iterate the EM algo until convergence
    while iterations>0:
        print("it: ",iterations,"mu: ",mu[0],"sigma: ",sigma[1])
        #E - step 
        
        #calculate the responsibility matrix 
        for k in range(k):
            R[:, k] = w[k] * P(mu[k], sigma[k])
        
        
        log_likelihood = np.sum(np.log(np.sum(R, axis = 1)))
        log_likelihoods.append(log_likelihood)
        R = (R.T / np.sum(R, axis = 1)).T
        N_ks = np.sum(R, axis = 0)

        #M - step
        for k in range(k):
            # update means
            mu[k] = 1. / N_ks[k] * np.sum(R[:, k] * x.T, axis = 1).T
            x_mu = np.matrix(x - mu[k])

            # update covariances
            sigma[k] = np.array(1 / N_ks[k] * np.dot(np.multiply(x_mu.T,  R[:, k]), x_mu))

            #compute the weight
            w[k] = 1. / n * N_ks[k]
            print("k:",k)
        # if convergence break 
        iterations=iterations-1
        if len(log_likelihoods) < 2 : continue
        if np.abs(log_likelihood - log_likelihoods[-2]) < 0.0001: break
    return initial_mu,initial_sigma,mu,sigma
    
    
    
```

```python
ini_mu,ini_sigma,mu,sigma = gmm1(x)
```

### Visualize the EM algo

```python
import matplotlib.mlab as mlab
def plot_contours(data, means, covs, title):
    plt.figure()
    plt.plot([x[0] for x in data], [y[1] for y in data],'ko') # data
    
    delta = 0.025
    k = len(means)
    print(k)
    x = np.arange(-2.0, 25.0, delta)
    y = np.arange(-2.0, 12.0, delta)
    X, Y = np.meshgrid(x, y)
    col = ['green', 'red', 'indigo','blue','yellow']
    for i in range(k):
        mean = means[i]
        cov = covs[i]
        sigmax = np.sqrt(cov[0][0])
        sigmay = np.sqrt(cov[1][1])
        sigmaxy = cov[0][1]/(sigmax*sigmay)
        Z = mlab.bivariate_normal(X, Y, sigmax, sigmay, mean[0], mean[1], sigmaxy)
        plt.contour(X, Y, Z, colors = col[i])
        plt.title(title)
    plt.rcParams.update({'font.size':16})
    plt.tight_layout()
```

**Plot the graph before and after fitting the model**

```python
plot_contours(x, ini_mu, ini_sigma, 'Initial clusters')
```

```python
plot_contours(x, mu, sigma, 'final clusters')
```
